### Step 2:  Pull in Data from API

This step utilizes the ONet Codes to obtain information from the API. This information that is obtained pulls from the API into lists.

```python
# Save baseurl, userID, state & API Token
baseurl = 'https://api.careeronestop.org/v1/lmi/'
userID = 'X96H8YEpkmPKz41'
apitoken = 'nSzfP81iOdFTETZZZbG54U3rJ8ertdAkv3r+i8i2Yi62nFM2KTlB3tWXkn0R/vZeTrTEGXV6xqUZCmMI+zuLsg=='
location = 'NE'
```

```python
# Create empty lists for storing API responses
career_outlook = []
ave_pay_ne = []
ave_pay_national = []
typical_training = []
soc_code = []
soc_title = []
soc_description = []
```

```python
# Create a function for adding API responses to lists
def api_to_list(data):
    career_outlook.append(data['LMI']['CareerOutLook'])
    ave_pay_ne.append(data['LMI']['AveragePayState'])
    ave_pay_national.append(data['LMI']['AveragePayNational'])
    typical_training.append(data['LMI']['TypicalTraining'])
    soc_code.append(data['SocInfo']['SocCode'])
    soc_title.append(data['SocInfo']['SocTitle'])
    soc_description.append(data['SocInfo']['SocDescription'])
```

```python
# Create a function to search for the ONet Code
def find_onet(code):
    try:
        url = baseurl + userID + '/' + code + '/' + location
        print('Retrieving the data of code',  code ,'now... ')
        jsonData = requests.get(url, headers={'Authorization': 'Bearer ' +  apitoken}).json()
        api_to_list(jsonData)
    except:
        print('error')
```

```python
# Call API by looping through the list of ONet codes
for i in df_onet['O*NET-SOC Code']:
    find_onet(i)
```

### Transformation 1: Create a dataframe from lists

Once the API pulls completed, the data was contained in the various lists. Therefore, I needed to combine the lists into one dataframe.

```python
# Build the dataframe
df = pd.DataFrame(list(zip(career_outlook, ave_pay_ne, ave_pay_national, 
                           typical_training, soc_code, soc_title, soc_description)))

# View the dataframe
df.head()
```

```python
# Load library
import sqlite3
```

```python
# Build table
table = """ CREATE TABLE API7
            (Career_Outlook VARCHAR(30),
            Average_Pay_NE FLOAT,
            National_Average_Pay FLOAT,
            Typical_Training VARCHAR(30),
            SOC_Code INTEGER,
            Title VARCHAR(30),
            Description VARCHAR(300)
            ); """
```

```python
# Build connection
con = sqlite3.connect('mydata2.sqlite')
```

```python
# Add table to database
con.execute(table)
con.commit()
```

```python
# Convert dataframe to list
df_list = df.values.tolist()
```

```python
# Add data to sql table
stmt = "INSERT INTO API7 VALUES(?, ?, ?, ?, ?, ?, ?)"
con.executemany(stmt, df_list)
con.commit()
```

```python
# View the data in the table
cursor = con.cursor()
print("Contents of the table:")
cursor.execute('''SELECT * from API7''')
print(cursor.fetchall())
```

```python
# Close the connection
con.close()
```
