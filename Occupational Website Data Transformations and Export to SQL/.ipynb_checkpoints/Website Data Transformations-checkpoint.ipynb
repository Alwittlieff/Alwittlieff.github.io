### Step 1: Replace Headers

```python
# Remove click instructions on Occupation title heading
df_bls.rename(columns={'Occupation title (click on the occupation title to view its profile)' : 'Occupation title'},inplace=True)
```

```python
# Verify heading update successful
df_bls.head()
```

```python
# Replace Occupation code with "SOC" to match flat file verbiage
df_bls.rename(columns={'Occupation code' : 'SOC'},inplace=True)
```

```python
# Verify new heading
df_bls.head()
```

For this step, I cleaned up the heading that featured instructions specific to the webpage. I also replaced one of the key identifying headings to align with the verbiage used in my flat file to ensure consistency. 

### Step 2: Remove Unnecessary Rows

```python
# Remove first row of data in DataFrame due to NaN values
df_bls = df_bls.drop(0)
df_bls
```

```python
# Remove last row of data in DataFrame due to NaN values
df_bls = df_bls.drop(1106)
df_bls
```

For my second step, I removed inconsistent values by deleting the rows. The two rows I removed were the first and last rows of data because both featured primarily NaN values.

### Step 3: Remove Dollar & Percent Signs

```python
# Received error when attempting to plot Employment RSE values
# Check data types
datatypes = df_bls.dtypes
datatypes
```

```python
# Remove dollar signs from the Median hourly wage field
df_bls['Median hourly wage'] = df_bls['Median hourly wage'].str.replace('$', '', regex=True)

# Remove dollar signs from the Mean hourly wage field
df_bls['Mean hourly wage'] = df_bls['Mean hourly wage'].str.replace('$', '', regex=True)

# Remove dollar signs from the Annual mean wage field
df_bls['Annual mean wage'] = df_bls['Annual mean wage'].str.replace('$', '', regex=True)
```

```python
# Validate changes
df_bls
```

```python
# Remove percentage signs from the Employment RSE field
df_bls['Employment RSE'] = df_bls['Employment RSE'].str.replace('%', '', regex=True)

# Remove dollar signs from the Mean wage RSE field
df_bls['Mean wage RSE'] = df_bls['Mean wage RSE'].str.replace('%', '', regex=True)
```

```python
# Validate changes
df_bls
```

This step took me quite a bit of time because I discovered that the data types were not what I had originally anticipated. Therefore, I needed to reformat the data. Originally, this step was going to be to check for bad data by ploting the values. However, the step became a vital data transformation to make the data more workable.  Multiple columns featured data containing dollar or percent signs. These dollar signs and percent signs needed to be removed to make the data usable. I then needed to move the check for bad data to a later step. 

### Step 4: Transform Data into Usable Format

The preceeding step did not make the data fully usable, as I could not plot values that are not yet in a numeric format.

```python
# Convert Employment RSE column to numeric type
df_bls['Employment RSE'] = df_bls['Employment RSE'].apply(pd.to_numeric)
```

```python
# Validate type change successful
datatypes = df_bls.dtypes
datatypes
```

```python
# Convert Employment RSE column to numeric type
df_bls['Median hourly wage'] = pd.to_numeric(df_bls['Median hourly wage'], errors ='coerce')
```

```python
# Convert Employment RSE column to numeric type
df_bls['Mean hourly wage'] = pd.to_numeric(df_bls['Mean hourly wage'], errors ='coerce')
```

```python
# Remove commas
df_bls['Annual mean wage'] = df_bls['Annual mean wage'].str.replace(',', '', regex=True)

# Convert Employment RSE column to numeric type
df_bls['Annual mean wage'] = pd.to_numeric(df_bls['Annual mean wage'], errors ='coerce')
```

```python
# Convert Employment RSE column to numeric type
df_bls['Mean wage RSE'] = pd.to_numeric(df_bls['Mean wage RSE'], errors ='coerce')
```

```python
# Validate type change successful
datatypes = df_bls.dtypes
datatypes
```

### Step 8: Export to SQL

```python
# Rename Columns to Avoid Duplicate Verbiage
df_bls.rename(columns={'Employment RSE' : 'EMPLY RSE', 'Employment per 1,000 jobs' : 'EMP per one thousand jobs', 'Mean hourly wage' : 'Hourly wage mean'},inplace=True)
```

```python
# Load library
import sqlite3
```

```python
# Build table
table = """ CREATE TABLE bls4
            (SOC VARCHAR(10),
            Occupation_title VARCHAR(300),
            Level VARCHAR(30),
            Employment FLOAT,
            EMPLY_RSE FLOAT,
            EMP_per_one_thousand_jobs FLOAT,
            Median_hourly_wage FLOAT,
            Hourly_wage_mean FLOAT,
            Annual_mean_wage FLOAT,
            Mean_wage_RSE FLOAT
            ); """
```

```python
# Build connection
con = sqlite3.connect('mydata2.sqlite')
```

```python
# Add table to database
con.execute(table)
con.commit()
```

```python
# Convert dataframe to list
df_list = df_bls.values.tolist()
```

```python
# Add data to sql table
stmt = "INSERT INTO bls4 VALUES(?, ?, ?, ?, ?, ?, ?, ?, ?, ?)"
con.executemany(stmt, df_list)
con.commit()
```

```python
# View the data in the table
cursor = con.cursor()
print("Contents of the table:")
cursor.execute('''SELECT * from bls4''')
print(cursor.fetchall())
```

```python
# Close the connection
con.close()
```
