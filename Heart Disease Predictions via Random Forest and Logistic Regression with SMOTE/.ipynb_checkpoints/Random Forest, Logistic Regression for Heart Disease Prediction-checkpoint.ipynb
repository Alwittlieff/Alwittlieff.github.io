# Data Preparation

#### Step 1: Create Dummy Variables

Dummy variables will allow me to leverage the survey variables that previously featured "yes" or "no" data in my analysis.

```python
# Leverage get dummies to create dummy variables across multiple columns
# Removes redundant columns
df_heart = pd.get_dummies(df_heart, drop_first=True)
df_heart
```

At this point, I do not want to remove any data because any of these risk factors could potentially correlate with whether an individual has heart disease. 

#### Step 2: Check that Data Type is Ready for Future Analysis

```python
# Check data types
datatypes = df_heart.dtypes
datatypes
```

The dummy variables show as "unit8" type. This stands for unsigned integers of 8 bits. Since this is a type of integer, I should be able to complete my data analysis without transforming the data to another type.  

#### Step 3: Check for Missing Data

```python
# Find missing values
# Leverage isnull functionality to view total missing values by header/column
df_heart.isnull().sum()
```

```python
# Get total null values
df_heart.isnull().sum().sum()
```

The results show no missing values in the dataset. Therefore, no action is required to handle NaN/missing values.

#### Step 4: Split into Training & Test Sets

```python
# Import library
from sklearn.model_selection import train_test_split
```

```python
# Split the data into a training and test set, where the HasHeartDisease column is the target
x = df_heart.drop('HasHeartDisease', axis=1)
y = df_heart['HasHeartDisease']
```

```python
# Create training and test sets with 80-20 split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

# Model Building and Evaluation

#### Step 1: Create a correlation coefficient matrix and visualization to get a feel for the correlations.

```python
# Import libraries
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Correlation matrix
correlation_mat = df_heart.corr()
```

```python
# Display correlation matrix
correlation_mat
```

```python
# Plot matrix using seaborn visualization library
sns.heatmap(correlation_mat)
plt.show()
```

#### Step 2: Check for Unbalanced Classes

```python
# Show the "no" (0) and "yes" (1) value counts for HasHeartDisease column
df_heart['HasHeartDisease'].value_counts()
```

Based on the information above, I discovered that the classes in my dataset are not balanced. Only approximately 9% of the responses featured "yes" for has heart disease. This is a highly unequal distribution of the two classes. Therefore, the algorithm may get biased towards the majority class and thus tend to predict the output as the majority class. In this case, the majority class is "no" for has heart disease. 

#### Step 3: Oversample with SMOTE

To correct the imbalanced classes, I elected to leaverage synthetic minority oversampling technique or "SMOTE." This duplicates examples from the minority class in the training dataset prior to fitting the model. Thus, the class distribution can be balanced without providing any additional information to the model.

```python
#pip install -U imbalanced-learn
```

```python
#pip install -U scikit-learn
```

```python
# Load libraries
import imblearn
from collections import Counter
from imblearn.over_sampling import SMOTE
```

```python
# Define oversampling strategy
SMOTE = SMOTE()
```

```python
# Fit and apply the transform
x_train_SMOTE, y_train_SMOTE = SMOTE.fit_resample(x_train, y_train)
```

```python
# Summarize class distribution to verify SMOTE success
print("Oversampling Outcome: ",Counter(y_train_SMOTE))
```

#### Step 4: Train a logistic regression using the training data.

```python
# Import libraries
from sklearn.preprocessing import StandardScaler
```

```python
# Create standardizer
standardizer = StandardScaler()
```

```python
# Standardize features
x_train_s = standardizer.fit_transform(x_train_SMOTE)
x_test_s = standardizer.transform(x_test)
```

```python
# Load library
from sklearn.linear_model import LogisticRegression

# Create logistic regression object
logistic_regression = LogisticRegression(random_state=0)

# Train model and predict
y_test_pred = logistic_regression.fit(x_train_s, y_train_SMOTE).predict(x_test_s)
```

```python
# Load library
from sklearn.metrics import accuracy_score

# Calculate accuracy
accuracy_score(y_test, y_test_pred)
```

The logistic regression yielded an accuracy score of approximately 75%. 

#### Step 5: Create a confusion matrix for the test set predictions.

```python
# Load libraries
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Create confusion matrix
matrix = confusion_matrix(y_test, y_test_pred)

# Create pandas dataframe
df = pd.DataFrame(matrix)

# Create heatmap
sns.heatmap(df, annot=True, fmt="d", cbar=None, cmap="Blues")
plt.title("Logistic Regression Model Confusion Matrix"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

For the model's purpose of predicting heart disease, the key area to review on the Logistic Regression Model Confusion Matrix is the lower left section. This represents the instances of individuals truly having heart disease when the model predicts that they would not have heart disease. These individuals could be missed due to not being flagged for further monitoring or action. 

#### Step 6: Get the precision, recall, and F1-score for the test set predictions.

```python
# Load libraries
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate Precision
# Precision Definition: Proportion of every observation predicted + that is actually +, measurement of noise in predictions
p = "{:.0%}".format(precision_score(y_test, y_test_pred))

# Calculate Recall
# Recall Definition: Proportion of every + observation that is truly +, model's ability to ID obervation of + class
r = "{:.0%}".format(recall_score(y_test, y_test_pred))

# Calculate F1-score
# F1 Definition: Measure of correctness achieved in + prediction, of obervations labeled +, how many actually +
f1 = "{:.0%}".format(f1_score(y_test, y_test_pred))

# Print Evaluation Metrics
print("Precision:",p)
print("Recall:",r)
print("F1-score:",f1)
```

#### Step 7: Build a Random Forest Model

```python
# Load library
from sklearn.ensemble import RandomForestClassifier
```

```python
# Build Random Forest Model
rf = RandomForestClassifier(n_estimators=1000, random_state=42)

# Train the model on training data
rf.fit(x_train_SMOTE, y_train_SMOTE)
```

```python
# Build predictions
rf_pred = rf.predict(x_test)
```

```python
# Calculate accuracy
accuracy_score(y_test, rf_pred)
```

The Random Forest Model achieved an accuracy score of 86%. This value is higher than the accuracy score of the Logistic Regression Model (approx. 75%). 

#### Step 8: Create a confusion matrix for the test set predictions.

```python
# Create confusion matrix
rf_matrix = confusion_matrix(y_test, rf_pred)

# Create pandas dataframe
rf_df = pd.DataFrame(rf_matrix)

# Create heatmap
sns.heatmap(rf_df, annot=True, fmt="d", cbar=None, cmap="Blues")
plt.title("Random Forest Model Confusion Matrix"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.show()
```

According to the Random Forest Model Confusion Matrix, the Random Forest Model correctly predicts those that do not have Heart Disease at a higher rate than the Logistic Regression Model. However, the Random Forest Model also more frequently misses the key area of individuals not being flagged for Heart Disease who should have been. 

#### Step 9: Get the precision, recall, and F1-score for the test set predictions.

```python
# Load libraries
from sklearn.metrics import precision_score, recall_score, f1_score

# Calculate Precision
# Precision Definition: Proportion of every observation predicted + that is actually +, measurement of noise in predictions
p = "{:.0%}".format(precision_score(y_test, rf_pred))

# Calculate Recall
# Recall Definition: Proportion of every + observation that is truly +, model's ability to ID obervation of + class
r = "{:.0%}".format(recall_score(y_test, rf_pred))

# Calculate F1-score
# F1 Definition: Measure of correctness achieved in + prediction, of obervations labeled +, how many actually +
f1 = "{:.0%}".format(f1_score(y_test, rf_pred))

# Print Evaluation Metrics
print("Precision:",p)
print("Recall:",r)
print("F1-score:",f1)
```
